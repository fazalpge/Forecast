```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import statsmodels.stats.api as sm
import math

%matplotlib inline
%pylab inline



# 1

#    ,      [a;b]
a=1
b=100

uniform_rv=sts.uniform(a,b-a)

ar=uniform_rv.rvs(1000)

x=np.linspace(a,b,1000)
pdf1 = uniform_rv.pdf(x)

# 
plt.figure(figsize(8,6))
plt.hist(ar,bins = 20, normed = True)
plt.plot(x,pdf1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

# 2


#       10
z10 = np.zeros(1000) 
for n in arange(1000):
     z10[n]=sum(uniform_rv.rvs(10))/10. #   10     
    
#   

#
EX=(b+a)/2.
#
DX=(b-a)**2/12.

#  
mu = EX
sigma = math.sqrt(DX/10.)

#     
norm_rv = sts.norm(loc=mu, scale=sigma)

#  1000 
norm_rv.rvs(size=1000)

#  
pdf = norm_rv.pdf(x)    
    
#     
    
plt.figure(figsize(8,6))
plt.hist(z10,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')


# 3

#       50
z50 = np.zeros(1000) 
for n in arange(1000):
     z50[n]=sum(uniform_rv.rvs(50))/50. #   10     
        
  #   

#
EX=(b+a)/2.
#
DX=(b-a)**2/12.

#  
mu = EX
sigma = math.sqrt(DX/50.)

#     
norm_rv = sts.norm(loc=mu, scale=sigma)

#  1000 
norm_rv.rvs(size=1000)

#  
pdf = norm_rv.pdf(x)          
#     
plt.figure(figsize(8,6))
plt.hist(z50,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')


# 4

#       300
z300 = np.zeros(1000) 
for n in arange(1000):
     z300[n]=sum(uniform_rv.rvs(300))/300. #   10     
        
  #   

#
EX=(b+a)/2.
#
DX=(b-a)**2/12.

#  
mu = EX
sigma = math.sqrt(DX/300.)

#     
norm_rv = sts.norm(loc=mu, scale=sigma)

#  1000 
norm_rv.rvs(size=1000)

#  
pdf = norm_rv.pdf(x)          
#     
plt.figure(figsize(8,6))
plt.hist(z300,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

#:     n,  
#       . 
#        n . 
```
